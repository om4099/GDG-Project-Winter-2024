```python
import requests
from bs4 import BeautifulSoup
import time

def scrape_poems(url, num_pages):
    for page in range(num_pages):
        print(f"Scraping page {page + 1}...")
        url = f"https://poets.org/poems?page={page}"
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            poems = []
            rows = soup.find_all('tr')
            if not rows:
                print("No articles found with the specified class.")
                return
            for row in rows:
                title_cell = row.find('td', class_="views-field views-field-title")
                poet_cell = row.find('td', class_="views-field views-field-field-author")
                year_cell =  row.find('td', class_="views-field views-field-field-date-published")
                if title_cell and poet_cell:

                    title = title_cell.find('a').text.strip()
                    poem_url = "https://poets.org" + title_cell.a['href']
                    poet = poet_cell.find('a').text.strip()
                    poet_url = "https://poets.org" + poet_cell.a['href']
                    year = year_cell.find('time').text.strip() if year_cell.find('time') else "Unknown"

                    print(f"Title: {title}, Poem URL: {poem_url}, Poet: {poet}, Poet URL: {poet_url}, Year: {year}")
                    poems.append([title, poem_url, poet, poet_url, year])

                # Uncomment the line below during debugging if needed
                # time.sleep(1)
        else:
            print(f"Failed to retrieve page: {response.status_code}")
#no of pages
num_pages = 5
scrape_poems("https://poets.org/poems?page=", num_pages)
```
